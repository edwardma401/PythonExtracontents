```python
import os
import pandas as pd
```

```python
file1 = "Resources/employee_data.csv"
file1_df = pd.read_csv(file1, encoding = "ISO-8859-1")
file1_df.head()
```

```python
new = file1_df["Name"].str.split(" ", n = 1, expand = True)
file1_df["First Name"] = new[0]
file1_df["Last Name"] = new[1]
file1_df.drop(columns = ["Name"], inplace = True)
file1_df.head()
```

```python
DOB_df = file1_df["DOB"]
DOB_df.head()
```

```python
new1 = DOB_df.str.split("-", n = 2, expand = True)
new1.head()
```

```python
new1 = new1[[1, 2, 0]]
new1.head()
```

```python
new2 = new1[1].map(str) + '/' + new1[2].map(str) + '/' + new1[0].map(str)
new2.head()
```

```python
file1_df.drop(columns = ["DOB"], inplace = True)
file1_df["DOB"] = new2
file1_df.head()
```
